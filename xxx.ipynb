```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("2025-09-02T12-54_export.csv")

# 12M 지표 다시 계산 (혹시 없을 경우 대비)
df["revenue_12m"] = df["MonthlyCharges"] * 12
df["expected_loss_12m"] = df["revenue_12m"] * df["churn_prob"]


import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(df["revenue_12m"], df["expected_loss_12m"], alpha=0.6)

# y=x 기준선
max_val = df["revenue_12m"].max()
plt.plot([0, max_val], [0, max_val], "r--", label="y=x (매출=손실)")

plt.xlabel("정상적 매출 (12M Revenue)")
plt.ylabel("이탈 위험 매출 (Expected Loss 12M)")
plt.title("정상 매출 vs 1년 이탈 위험 매출")
plt.legend()
plt.show()
```
